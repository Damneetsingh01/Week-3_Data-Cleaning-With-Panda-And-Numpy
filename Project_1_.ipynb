* Importing Numpy and Panda In Google Colab

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/Data-cleaning-for-beginners-using-pandas.csv')
```

* Raw Data

```python
data
```

 * Copy the Data To preserve the original Data







```python
data2 = data.copy()
```

* Converting the columns name to lower case and removing the spaces

```python
data2.columns = data2.columns.str.lower().str.replace(" ","_")
```

* First 5 columns of Raw Data

```python
data2.head()
```

* Last 5 columns of raw Data

```python
data2.tail()
```

* Getting The Shape

```python
data2.shape
```

* Getting The Info About the Data

```python
data2.info()
```

* Finding The null Values and Sum of them

```python
data2.isnull().sum()
```

* Find The Average Age

```python
avg_age = data2['age'].mean()
avg_age
```

* Replacing The Null Values in Age column with Average Age

```python
data2['age']=data2.age.fillna(avg_age).round().astype(int)
data2
```

* Recheck The Data Info

```python
data2.info()
```

* Correcting The Salary description

```python
data2['salary']=data2['salary'].str.replace("$"," ").str.replace("k","000")
data2
```

* Splitting The Salary Into Max Salary And Min Salary

```python
data2[['max salary','min salary']] = data2['salary'].str.split('-',expand = True)

# Coverting The Max and Min Salary into interger Type
data2['max salary'] = data2['max salary'].astype(int)
data2['min salary'] = data2['min salary'].astype(int)

# Dropping The salary column
data2 = data2.drop(columns=['salary'])

data2
```

* Info Of data

```python
data2.info()
```

* Count of -1.0 present in rating column

```python
count_of_minus_one = (data2['rating']== -1.0).sum()
count_of_minus_one
```

* Replacing Them

```python
data2['rating'] = data2['rating'].replace(-1.0,0.0)
data2.head()
```

* Count oF -1.0 present in established column

```python
count_of_minus_one_established = (data2['established'] == -1.0).sum()
count_of_minus_one_established
```

* Find The most Occuring Value in established

```python
mode_established = data2['established'].mode().iloc[1]
mode_established
```

* Replace the -1.0 with most occuring value

```python
data2['established'] = data2['established'].replace(-1.0,mode_established)
data2.head()
```

* Formatting String in 'location' column

```python
data2['location'] = data2['location'].replace({'New York,Ny':'New York', 'India,In':'India','India In':'India', 'Australia Aus':'Australia'})
data2.head()
```

* Converting datatype of 'location' column from object to string

```python
data2['location'] = data2['location'].astype('string')
data2.info()
```

* Finding the unique values in 'easy_apply' column

```python
unique_values_count = data2['easy_apply'].unique()
unique_values_count
```

* Replace Them With appropriate results

```python
data2['easy_apply'] = data2['easy_apply'].replace('TRUE', True)

data2['easy_apply'] = data2['easy_apply'].replace('-1', False)

data2['easy_apply'] = data2['easy_apply'].astype(bool)

data2.info()

data2.head()
```

* Replacing The Null Value in rating

```python
data2['rating'].isnull().sum()

# Get The Average of The Ratings
mean_rating=data2['rating'].mean()
mean_rating

#Replacing the values
data2['rating'] = data2.rating.fillna(mean_rating).round(decimals=1)
data2.head()
```
